```python
from mlx_lm import load, generate
```

```python
def generate_output(model, tokenizer, prompt, max_tokens=2048, verbose=True):
    response = generate(
        model,
        tokenizer,
        prompt=f"<|user|>\n {prompt} <|end|>\n<|assistant|>",
        max_tokens=max_tokens,
        verbose=verbose
    )
    return response
```

```python
model_id = "microsoft/Phi-3-mini-4k-instruct"
model, tokenizer = load(model_id)
```

```python
prompt = "<|user|>\n Create a funny joke about chickens.<|end|>\n<|assistant|>"
```

```python
response = generate_output(
    model,
    tokenizer,
    prompt=f"<|user|>\n {prompt} <|end|>\n<|assistant|>"
)
print(response)
```
